```python
import random
```

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
def code_to_debug():
  import pdb; pdb.set_trace()

  for i in range(10):
    x = random.random()
    factorial(x)
```
