- functions

```python
def print_squares(stop_val):
  for i in range(stop_val):
    print(i, i**2)
```

```python
print_squares(5)
```

```python
def sum_squares(stop_val):
  total = 0
  for i in range(stop_val):
    total += i**2
  return total
```

```python
sum_squares(5)
```

```python
sum_5 = sum_squares(5)
```

```python
sum_5
```

```python
def print_squares_with_startval(stop_val,start_val=0):
  for i in range(start_val,stop_val):
    print(i, i**2)
```

```python
print_squares_with_startval(5,3)
```

# Python Classes

```python
class Person():
  """docstring for Person"""
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def __str__(self):
    return self.name
  def get_age(self):
    return self.age
```

```python
new_person = Person("John", 30)
```

```python


type(new_person)
```

```python
new_person.get_age()
```

```python

new_person.name
```

```python
Person.__doc__
```

```python
class iPhone(Person):
  """docstring for iPhone"""
  def __init__(self, name, age, model):
    super().__init__(name, age)
    self.model = model
  def get_model(self):
    return self.model
```

```python
iphone = iPhone("John", 30, "Xiaomi")
```

```python
iphone.get_model()
```
